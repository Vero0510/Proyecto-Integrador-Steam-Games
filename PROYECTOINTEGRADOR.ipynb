```python
import matplotlib.pyplot as plt
```

```python
total_positive_ratings = Juego_en_Steam['positive_ratings'].sum()
total_negative_ratings = Juego_en_Steam['negative_ratings'].sum()
```

```python
total_negative_ratings
```

```python
total_positive_ratings
```

```python
# Datos para el gráfico
labels = ['Positive Ratings', 'Negative Ratings']
sizes = [total_positive_ratings, total_negative_ratings]
colors = ['lightblue', 'lightcoral']
explode = (0.1, 0)  # Resaltar la primera porción (Positive Ratings)

# Crear el gráfico circular
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%',
        shadow=True, startangle=140)

# Aspecto igual para que sea circular
plt.axis('equal')

# Mostrar el gráfico
plt.title('Distribution of Positive and Negative Ratings')
plt.show()
```
